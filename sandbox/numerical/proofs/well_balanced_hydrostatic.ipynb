# Hydrostatic reconstruction + HLLC well-balanced proof

For `eta = const` and `hu = hv = 0`, every edge normal velocity is zero for triangle-triangle, quad-quad, quad-triangle, and boundary edges.

```python
import sympy as sp
h, phi_e_n, nx, ny = sp.symbols('h phi_e_n nx ny', positive=True)
hu = hv = sp.Integer(0)
u_n = (hu / h) * nx + (hv / h) * ny
F_h = phi_e_n * h * u_n
sp.simplify(F_h)
```

The result is independent of cell topology; only the edge normal projection appears. Therefore E1 quad-tri, E3 tri-tri, quad-only quad-quad, and boundary edges all have zero mass flux.
