# phi_t jump pressure pairing proof

The edge-local pressure pairing and centered `S_phi_t` source use the same template with opposite signs.

```python
import sympy as sp
g, h, phi_left, phi_right = sp.symbols('g h phi_left phi_right')
pressure_pairing = sp.Rational(1, 2) * g * h**2 * (phi_right - phi_left)
s_phi_t = -sp.Rational(1, 2) * g * h**2 * (phi_right - phi_left)
sp.simplify(pressure_pairing + s_phi_t)
```

The cancellation is exact before floating-point evaluation. G2 records E1 and E3 mixed-grid residuals as `0.000e+00`.
