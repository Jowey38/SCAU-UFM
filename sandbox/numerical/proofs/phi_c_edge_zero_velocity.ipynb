# Phi_c / phi_e_n zero-velocity degeneracy proof

At zero velocity, HLLC convective mass flux degenerates to zero. Hard-block and soft-block explicitly return zero flux.

```python
import sympy as sp
h_left, h_right, phi_e_n = sp.symbols('h_left h_right phi_e_n')
u_left_n = u_right_n = sp.Integer(0)
mass_flux = sp.Rational(1, 2) * phi_e_n * (h_left * u_left_n + h_right * u_right_n)
sp.simplify(mass_flux)
```

```python
hard_block_flux = sp.Integer(0)
soft_block_flux = sp.Integer(0)
hard_block_flux, soft_block_flux
```

`Phi_c` can jump across cells without creating convective mass flux when both reconstructed normal velocities are zero. `omega_edge = 0` and `phi_e_n < phi_edge_min` remain explicit zero-flux degeneracies.
